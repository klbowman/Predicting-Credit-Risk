```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df.head()
```

```python
# Convert categorical data to numeric for training data
train_df = pd.get_dummies(data=train_df, columns=['home_ownership','verification_status','loan_status','pymnt_plan','initial_list_status','application_type','hardship_flag','debt_settlement_flag'])
```

```python
# Verify that get_dummies() worked 
train_df.columns
```

```python
# Separate target feature for training data
y_train = train_df['loan_status_high_risk']
X_train = train_df.drop('loan_status_high_risk', axis=1)
X_train = X_train.drop('loan_status_low_risk',axis=1)
```

```python
# Convert categorical data to numeric for testing data
test_df = pd.get_dummies(data=test_df, columns=['home_ownership','verification_status','loan_status','pymnt_plan','initial_list_status','application_type','hardship_flag','debt_settlement_flag'])
```

```python
# Verify that get_dummies() worked 
test_df.columns
```

```python
# Separate target feature for testing data
y_test = test_df['loan_status_high_risk']
X_test = test_df.drop('loan_status_high_risk', axis=1)
X_test = X_test.drop('loan_status_low_risk',axis=1)
```

```python
# add missing dummy variables to testing set
# debt_settlement_flag_Y missing from testing set
X_test['debt_settlement_flag_Y']=0
X_test.head()
```

Prediction: The logistic regress model will perform better than the random forest classifer model. There are several categories in the dataset that likely have litte influence on the risk category, adding noise to the data. This noise will decrease the accuracy of the random forest classifier model.

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)
print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

My prediction was incorrect, the random forest classifier model performed better than the logistic regression model. 

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Prediction: Scaling will improve both model's scores, and the random forest classifier model will still have a higher score.

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier.fit(X_train_scaled, y_train)
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

Scaling the data improved the score of the logistic regression model, but did not affect the score of the random forest classifier model. The random forest classifier model may be scaling down the relevance of features that important for predicting credit risk.
